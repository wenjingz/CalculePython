## Fibonacci 

```python
# Suite de Fibonacci (ni arithmetique, ni geometrique) avec affectation parallèle
f1, f2 = 1, 1                     # f1 = f2 = 1
while f2 < 100:
    print(f2, end = " ")          # on affiche que le 2nd terme calcule parmi les 2
    f1, f2 = f2, f1 + f2          # f2 pour identifier fn+1 = fn + fn-1, f2 l'ancien fn-2, f1 l'ancien fn-1
                                  # f1 identifie fn et memorise fn-1
```

## Valuation dyadique

```python
# Valuation dyadique ... voir aussi plus loin pour la mise en oeuvre sous forme de fonction Python    
a = 400 # 100 # 31250 # 100
b = a
v = 0
while b % 2 == 0:
    b //= 2
    v += 1
print('Valuation dyadique de ',a, ' = ', v)
```

## Algorithme d'Euclide？

```python
a = 84
b = 72
while b != 0 :  # ou not b
    r = a % b
    a , b = b , r
print('pgcd(72,84) = ',a)
```

## Conjecture de Syracuse ... constatée mais jamais demontrée？

```python
n = 127    # commencer par 27, essayer avec 7 .. pour 127, le temps de vol = duree est de 46, altitude max = hauteur de 4372
duree = 0
hauteur = 0
while n > 1:
    duree += 1
    if n % 2:   # test n impair car n%2=0 => bool(0)=False <=> n impair
        n = 3 * n + 1  # Un+1 = 3*Un + 1
    else:       # n est pair, Un+1 = Un/2
        n //= 2
    if n > hauteur:  # on conserve la valeur max de Un
        hauteur = n
print('hauteur = ', int(hauteur),' Durée = ', duree) 
print('valeur finale = ',n)
```

```python
# voir les regles de construction (description des parametres E/S, usage, exemple)

def fonction1(x):     # bonnes pratiques : à placer en 2nd bloc d'un module
    y = 2 * x + 1
    return y

def fonction2(x):
    y = x * x
    return y

def fonction3(fonc, x):
    y = fonc(x)
    return y

print(fonction3(fonction1, 4))  # fonction comme parametre de fonction, on trouve 9 car f1(4)=2*4+1
print(fonction3(fonction2, 5))  # f2(5) = 25

def AirePerimetre(R):
    a = pi * R ** 2
    p = 2 * pi * R
    return a, p   # renvoie 2 paramètres de sortie
    
aire, peri = AirePerimetre(3)    
```

```python
# tester un appel partiel aire = ... puis editer peri
   
def Factorielle(n):
    f = 1
    for k in range(2,n + 1):  # rappel : on n atteint pas la borne sup
        f *= k
    return f

def FactorielleRecursive(n):    # 1er exemple d'un modèle récursif
    if n == 0 or n == 1:        # condition pour sortir de la récursivité
        return 1                
    else:
        return n * FactorielleRecursive(n - 1)

Factorielle(4)
Factorielle(5)
FactorielleRecursive(5)
```

```python
def produit_cartesien(A,n):
    if n == 0:
        return [[]]
    return [[a] + b for a in A for b in produit_cartesien(A, n - 1)]  # + pour la concatenation, les 2 boucles for pour les combinaisons dynamiques a=liste source
            
A=[1,2]
print('A=',A)
print()

for k in range(5):
    B=produit_cartesien(A,k)
    print('A^'+str(k)+'=',B)
    print()
```

```python
# Une suite qui a l'air décroissante à partir d'un certain rang, et convergente
# Evolution de la valeur des sommes partielles <=> Sigma des (k/n)^k pour k variant de 1 à n pour n donné

def Somme(n):
    S=0
    for k in range(1,n+1):
        S+=(k/n)**k
    return S
    
for n in range(1000001):   # tester avec 10^7 pour observer la lente decroissance sous la valeur seuil 1.5820
    if not n % 5000: # On ne va quand même pas tout afficher
        print('S(',n,')=',Somme(n))
```

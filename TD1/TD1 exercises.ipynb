## 1.4 Calcul de la factorielle d’un entier de manière itérative et récursive

### itérative

```python
def Fact(n):
  result = n
  for i in range(1,n):
    result*=i
  return result

print(Fact(6))
```

### récursive

```python
def FactR(n):
  if n == 1:
    return 1
  else:
    return n*FactR(n-1)

print(FactR(6))
```

## 1.5 Comparaison de deux listes

```python
def Compare(A,B):
  nA = len(A)
  nB =len(B)
  C = [True,True,True]
  if nA != nB:
    print('Comparaison impossible')
  else:
    for i in range(nA):
        if A[i]!=B[i]:
            C[i]=False
  return C
      
A=['a',[1,0],np.pi]
B= ['a',[1,2],7]
Compare(A,B)
  
```

### Exercise 9

```python
def Sym1(A):
    [n,p] = np.shape(A) #Pour la dimension de A
    for i in range(0,n):
        for j in range(0,p):
            if A[i,j]!=A[j,i]:
                return False
    return True

A=np.array([[2,1,1],[1,2,1],[1,1,2]])
B=np.array([[2,1,0],[1,2,1],[1,1,2]])

print(Sym1(A))
print(Sym1(B))
```

### Exercice 10

```python
def Sym2(A):
    B = np.transpose(A)
    if (A==B).all():
        return True
    return False

A=np.array([[2,1,1],[1,2,1],[1,1,2]])
B=np.array([[2,1,0],[1,2,1],[1,1,2]])

print(Sym2(A))
print(Sym2(B))
```

## 2.2 Construction de matrices

### Exercice 11

```python
def Tridiag(a,b,c,n):
    array1 = np.diag([a]*n,0)
    array2 = np.diag([b]*(n-1),1)
    array3 = np.diag([c]*(n-2),2)
    
    return array1+array2+np.transpose(array2)+array3+np.transpose(array3)
Tridiag(1,2,3,5)
```

```python
def Tridiag(a,b,c,n):
    A = np.diag( a*np.ones(n) ) + np.diag(b*np.ones(n-1),1) + np.diag(b*np.ones(n-1),-1)\
        + np.diag(c*np.ones(n-2),2) + np.diag(c*np.ones(n-2),-2)
    return A

Tridiag(1,2,3,5)
```

## 2.3 Produit matriciel

### Exercice 13.

```python
def ProduitMatriceParColonne(M,X):
    [n,p]= np.shape(M)
    [m,q]=np.shape(X)
    if p!=m:
        print('Produit impossible')
    else:
        Y = np.zeros((n,q))#Matrice nulle
        for i in range(n):
            #Utilisation de la fonction native sum pour faire une somme
            Y[i]=sum( M[i][k]*X[k][i] for k in range(p))
    return Y

M=MatriceParticuliere(4)
X=np.ones((4,1))

ProduitMatriceParColonne(M,X)
```

## 2.4 Somme de deux polynômes

```python
def SommePoly(P,Q):
    n = len(P)
    m = len(Q)
    if n == m:
        return sum(P,Q)
    if n<m:
        return np.concatenate((Q,P))
    if n>m:
        return np.concatenate((P,Q))
    
P=np.array([1, 2, 3])
Q=np.array([1, 2 ])
SommePoly(P,Q)
```

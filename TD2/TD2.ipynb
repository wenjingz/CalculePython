```python
from math import *
import numpy as np
import math
import matplotlib.pyplot as plt


def Bessel1(n, x):
    s = 0
    for k in range(n):
        for i in range(3000):
            s+= ((-1)**k/(factorial(k)**2))*((x/2)**((2)*k))
    return s

def GraphBessel1(n, a, b):
    # Grille d'abscisses (3000 noeuds de calculs, arbitraire) 
    x = np.linspace(a, b, 3000)    
    # Calcul des images 
    y = Bessel1(n, x) 
    # Graphique 
    plt.plot(x, y) 
    plt.title('Bessel_'+str(n)+' sur ['+str(a)+', '+str(b)+']') 
    plt.show()
 
GraphBessel1(50,0,35)
GraphBessel1(50,0,30)
```
